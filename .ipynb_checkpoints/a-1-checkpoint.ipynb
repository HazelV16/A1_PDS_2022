```python
import pandas as pd
```

```python
# Load Engine CC csv file
engine_cc_filename = 'Engine CC.csv'
engine_cc = pd.read_csv(engine_cc_filename, sep = ',',decimal='.',header= 0, index_col = 0)
engine_cc
```

```python
# Load Female csv file
female_filename = 'Female.csv'
female = pd.read_csv(female_filename, sep = ',',decimal='.',header= 0, index_col = 0)
female
```

```python
# Load Fuel csv file
fuel_filename = 'Fuel.csv'
fuel = pd.read_csv(fuel_filename, sep = ',',decimal='.',header= 0, index_col = 0)
fuel
```

```python
fuel.iloc[23, 0]
```

```python
# Load male csv file
male_filename = 'Male.csv'
male = pd.read_csv(male_filename, sep = ',',decimal='.',header= 0, index_col = 0)
male
```

```python
# Load manufacturer csv file
manufacturer_filename = 'Manufacturer.csv'
manufacturer = pd.read_csv(manufacturer_filename, sep = ',',decimal='.',header= 0, index_col = 0)
manufacturer
```

```python
# Load model csv file
model_filename = 'Model.csv'
model = pd.read_csv(model_filename, sep = ',',decimal='.',header= 0, index_col = 0)
model
```

```python
# Load power csv file
power_filename = 'Power.csv'
power = pd.read_csv(power_filename, sep = ',',decimal='.',header= 0, index_col = 0)
power
```

```python
# Load price csv file
price_filename = 'Price.csv'
price = pd.read_csv(price_filename, sep = ',',decimal='.',header= 0, index_col = 0)
price
```

```python
# Load total csv file
total_filename = 'Total.csv'
total = pd.read_csv(total_filename, sep = ',',decimal='.',header= 0, index_col = 0)
total
```

```python
# Load transmission csv file
transmission_filename = 'Transmission.csv'
transmission = pd.read_csv(transmission_filename, sep = ',',decimal='.',header= 0, index_col = 0)
transmission
```

```python
# Load unknown csv file
unknown_filename = 'Unknown.csv'
unknown = pd.read_csv(unknown_filename, sep = ',',decimal='.',header= 0, index_col = 0)
unknown
```

```python
# check the types of each feature
print(engine_cc.dtypes)
print(female.dtypes)
print(fuel.dtypes)
print(male.dtypes)
print(manufacturer.dtypes)
print(model.dtypes)
print(power.dtypes)
print(price.dtypes)
print(total.dtypes)
print(transmission.dtypes)
print(unknown.dtypes)
```

```python
# merge all the csv files
merged_df = engine_cc.join(fuel).join(manufacturer).join(model).join(power).join(price).join(transmission).join(female).join(male).join(unknown).join(total)
merged_df
```

```python
# show the first 30 rows in the merged file
merged_df['Model'].value_counts()
```

```python
# show the total number of each manufacturer
merged_df.sort_values(['Engine CC'], ascending=False)
```

```python
# count the number of each fuel
merged_df['Fuel'].value_counts()
```

```python
# count the number of each Model
merged_df['Model'].value_counts() # need to check carefully 
```

```python
# drop nan value
merged_df_dropped = merged_df.dropna().reset_index(drop=True)
merged_df_dropped
```

```python
merged_df_dropped['Model'] = merged_df_dropped['Model'].str.strip()
merged_df_dropped['Model'].value_counts().tail(50)
# merged_df_dropped.sort_values(by = 'Model', ascending=False)
```

```python
# remove redundant whitespaces
merged_df_dropped['Manufacturer'] = merged_df_dropped['Manufacturer'].str.strip()
merged_df_dropped['Manufacturer'].value_counts()
```

```python
# replace peatrol w petrol
# replace diasel w diesel
# replace autometic w automatic
merged_df_dropped['Fuel'].replace('peatrol', 'petrol', inplace = True)
merged_df_dropped['Fuel'].replace('diasel', 'diesel', inplace = True)
merged_df_dropped['Fuel'].replace('autometic', 'automatic', inplace = True)
merged_df_dropped['Fuel'].value_counts()
```

```python
# change types from str to int
merged_df_dropped['Female'] = merged_df_dropped['Female'].astype('str').str.replace(',', '').astype('int')
merged_df_dropped['Male'] = merged_df_dropped['Male'].astype('str').str.replace(',', '').astype('int')
merged_df_dropped['Total'] = merged_df_dropped['Total'].astype('str').str.replace(',', '').astype('int')
merged_df_dropped['Unknown'] = merged_df_dropped['Unknown'].astype('str').str.replace(',', '').astype('int')
merged_df_dropped.dtypes
```

```python
# group by female male total and unknown
merged_df_dropped[['Female', 'Male', 'Total', 'Unknown']] = merged_df_dropped.groupby(['Engine CC', 'Fuel', 'Manufacturer', 'Model', 'Power', 
                                                  'Price', 'Transmission'])[['Female', 'Male', 'Total', 'Unknown']].transform('sum') 
merged_df_dropped[['Female', 'Male', 'Total', 'Unknown']]
```

```python
# remove duplicate
merged_df_dropped = merged_df_dropped.drop_duplicates(subset=['Engine CC', 'Fuel', 'Manufacturer', 'Model', 'Power', 
                                              'Price', 'Transmission']).reset_index(drop=True)
merged_df_dropped
```

```python
merged_df_dropped.sort_values(['Engine CC'], ascending=False).head(60)
```

```python
merged_df_dropped.to_csv('cleaned car buyers.csv', index = True)
```
